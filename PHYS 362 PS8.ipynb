```python
import random
import math
import matplotlib.pyplot as plt
```

```python
N=100

for i in range(15, -1, -1):
    kbT=0.25+0.25*i #Choosing the dimensionless Jeps/kbT, and setting J=1.
    grid=[[random.choice([1,-1])for i in range(N)] for j in range(N)] #Initalize the grid.

    for i in range (0,N*1000):
        randx=random.randrange(0,N)
        randy=random.randrange(0,N)

        epsi=-((grid[randx][randy])*(grid[(randx+1) % N][randy]+grid[(randx-1) % N][randy]+grid[randx][(randy+1) % N]
                   +grid[randx][(randy-1) % N])) #Initial energy.
        epsf=-((-grid[randx][randy])*(grid[(randx+1) % N][randy]+grid[(randx-1) % N][randy]+grid[randx][(randy+1) % N]
                   +grid[randx][(randy-1) % N])) #Initial energy.

        deltaeps=epsf-epsi #Delta epsilon.
        if (deltaeps<=0): #Delta epsilon less than 0, always flip.
            grid[randx][randy]=-grid[randx][randy]
        else:
            boltzfac=math.exp(-deltaeps/kbT)
            x=random.random() #Random number between 0 and 1
            if x<boltzfac: #This is just a way to implement probabilities.
                grid[randx][randy]=-grid[randx][randy] #Flip.
            else:
                grid[randx][randy]=+grid[randx][randy] #Don't flip.

    cmap = plt.cm.binary.reversed() #From ChatGPT

    plt.imshow(grid, cmap=cmap)
    plt.title("Magnetization Grid at kbT="+ str(kbT)+"(In Units of J)")
    plt.show() ##Showing the plot
```
